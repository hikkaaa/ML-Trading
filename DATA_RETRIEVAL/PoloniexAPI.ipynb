```python
# POLONIEX PRICES

prices = requests.get("https://api.poloniex.com/markets/price")
prices_list = []
if prices.status_code == 200:
  prices_json = json.loads(prices.text)
  for p in prices_json:
    prices_list.append({'ticker': p['symbol'], 'price': p['price']})
```

```python
# POLONIEX ORDERBOOK

# The term order book refers to an electronic list of buy and sell orders for a specific security or financial instrument organized by price level. 
# An order book lists the number of shares being bid on or offered at each price point, or market depth.
ticker = "DASH_BTC"
limit = 20
# ? - symbol to start the query
ob = requests.get(f"https://api.poloniex.com/markets/{ticker}/orderBook?limit={limit}")
ob_json = []
if ob.status_code == 200:
  ob_json = json.loads(ob.text)
print(ob_json)
```

```python
# structure our orderbook (for the 'asks')
asks_prices = []
asks_sizes = []
counts = 0

for a in ob_json["asks"]:
  if counts % 2 == 0:
    asks_prices.append(a)
  else:
    asks_sizes.append(a)
  counts += 1
print(asks_prices)
print(asks_sizes)
```
